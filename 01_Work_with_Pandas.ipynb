# Работа с библиотекой Pandas

1. Загрузка набора данных из файла **car_data.csv**



```python
import pandas as pd
!gdown 1kfwKDsUs9I3jaYxLQhbz8ofEIEHSGPBT
df = pd.read_csv('https://drive.google.com/uc?id=1kfwKDsUs9I3jaYxLQhbz8ofEIEHSGPBT')
```

2. Анализ пропусков в данных. Удаление строк или столбцов, содержащих пропуски.

```python
df = df.rename(columns={'Unnamed: 0': 'Unnamed:_0', 'Car_id': 'Car_id', 'Date': 'Date','Customer Name': 'Customer_Name', 'Gender': 'Gender',
       'Annual Income': 'Annual_Income', 'Dealer_Name': 'Dealer_Name', 'Company': 'Company', 'Model': 'Model', 'Engine': 'Engine',
       'Transmission': 'Transmission', 'Color': 'Color', 'Price ($)': 'Price_($)', 'Dealer_No ': 'Dealer_No', 'Body Style': 'Body_Style',
       'Phone': 'Phone', 'Dealer_Region': 'Dealer_Region'})
df.isnull().sum()
df.dropna(inplace=True)
```

3. Вычисление среднего дохода клиентов, купивших автомобиль марки Toyota Camry.

```python
camry_model = df[(df['Company'] == 'Toyota') & (df['Model'] == 'Camry')]

average_clients_income = camry_model['Annual_Income'].mean()
average_clients_income
```

4. Топ 10 самых дорогих марок автомобилей за представленный период времени.


```python
avg_prices = df.groupby('Company')['Price_($)'].mean().sort_values(ascending=False).head(10)
avg_prices
```

5. Средняя стоимость автомобилей по типам кузова (Body Style), используя метод groupby.

```python
average_cost_body_style = df.groupby('Body_Style')['Price_($)'].mean().sort_values(ascending=False)
average_cost_body_style
```

6. Автомобили, которые приобрели 10 клиентов с самым высоким доходом

```python
top_10_clients = df.sort_values('Annual_Income', ascending=False).head(10)
top_10_clients[['Customer_Name', 'Annual_Income', 'Company', 'Model']]
```

7. Автомобили, которые приобрели 10 клиентов с самым низким доходом


```python
low_10_clients = df.sort_values('Annual_Income', ascending=True).head(10)
low_10_clients[['Customer_Name', 'Annual_Income', 'Company', 'Model']]
```

8. Создание нового столбца, значения которого равны логарифму от величины дохода клиентов.


```python
import numpy as np
df['Log_Annual_Income'] = np.log(df['Annual_Income'])
df.head()
```
